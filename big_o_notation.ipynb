Curves to graph
- y = 0n + 1 and label the curve "O(1)"
- y = log(n) and label the curve "O(log n)"
- y = n and label the curve "O(n)"
- y = nlog(n) and label it "O(n log n)"
- y = n^2 and label it "O(n^2)"
- y = 2^n and label it "O(2^n)"
- y = n! and label it "O(n!)"
- y = n^n and label it "O(n^n)"

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from math import log, factorial

plt.figure(figsize = (10, 10))
n = list(range(0, 101))

curves = [
    {'y' : [1 for _ in n], 'c' : 'red', 'label' : 'O(1)'},
    {'y' : [log(x) if x != 0 else log(0.001) for x in n], 'c' : 'blue', 'label' : '$O(\lg n)$'},
    {'y' : [x for x in n], 'c' : 'green', 'label' : 'O(n)'},
    {'y' : [x * log(x) if x != 0 else x * log(0.001) for x in n], 'c' : 'orange', 'label' : '$O(n \lg n)$'},
    {'y' : [x * x for x in n], 'c' : 'purple', 'label' : '$O(n^2)$'},
    {'y' : [2 ** x for x in n], 'c' : 'pink', 'label' : '$O(2^n)$'},
    {'y' : [factorial(x) for x in n], 'c' : 'black', 'label' : 'O(n!)'},
    {'y' : [x ** x for x in n], 'c' : 'cyan', 'label' : '$O(n^n)$'}
]

for curve in curves:
    plt.plot(n, curve['y'], c = curve['c'], label = curve['label'])

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.xlim(0, 100)
plt.ylim(0, 500)
plt.legend(loc = 'center right');
```
