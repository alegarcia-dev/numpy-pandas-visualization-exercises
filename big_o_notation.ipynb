Curves to graph
- y = 0n + 1 and label the curve "O(1)"
- y = log(n) and label the curve "O(log n)"
- y = n and label the curve "O(n)"
- y = nlog(n) and label it "O(n log n)"
- y = n^2 and label it "O(n^2)"
- y = 2^n and label it "O(2^n)"
- y = n! and label it "O(n!)"
- y = n^n and label it "O(n^n)"

```python
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import linspace, log
from scipy.special import factorial
```

```python
x_min, x_max = 1, 100
y_min, y_max = 0, 500

plt.figure(figsize = (8, 8))

n = linspace(x_min, x_max + 1, x_max - x_min + 1)

curves = [
    {'y' : [1 for _ in n], 'c' : 'red', 'label' : 'O(1)'},
    {'y' : log(n), 'c' : 'blue', 'label' : '$O(\lg n)$'},
    {'y' : n, 'c' : 'green', 'label' : 'O(n)'},
    {'y' : n * log(n), 'c' : 'orange', 'label' : '$O(n \lg n)$'},
    {'y' : n * n, 'c' : 'purple', 'label' : '$O(n^2)$'},
    {'y' : 2 ** n, 'c' : 'pink', 'label' : '$O(2^n)$'},
    {'y' : factorial(n), 'c' : 'black', 'label' : 'O(n!)'},
    {'y' : n ** n, 'c' : 'cyan', 'label' : '$O(n^n)$'}
]

for curve in curves:
    plt.plot(n, curve['y'], c = curve['c'], label = curve['label'])

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.legend(loc = 'center right');
```
