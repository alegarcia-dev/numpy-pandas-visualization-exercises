```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from numpy import linspace
import math
```

# 1
Use matplotlib to plot the following equation:

<img src="https://latex.codecogs.com/svg.image?y&space;=&space;x^{2}&space;-&space;x&space;&plus;&space;2" title="y = x^{2} - x + 2" />

Add an anotation for the point (0, 0), the origin.

```python
x = list(range(-50, 51))
y = [n * n - n + 2 for n in x]

plt.plot(x, y)
plt.annotate('Origin', xy = (0, 0), xytext = (-20, 650), arrowprops = {'facecolor' : 'green'})
plt.title('$y = x^2 - x + 2$');
```

# 2
Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
- <img src="https://latex.codecogs.com/svg.image?y&space;=&space;\sqrt{x}" title="y = \sqrt{x}" />
- <img src="https://latex.codecogs.com/svg.image?y&space;=&space;x^{3}" title="y = x^{3}" />
- <img src="https://latex.codecogs.com/svg.image?y&space;=&space;2^{x}" title="y = 2^{x}" />
- <img src="https://latex.codecogs.com/svg.image?y&space;=&space;\frac{1}{x&plus;1}" title="y = \frac{1}{x+1}" />

```python
x = list(range(0, 21))
y = [math.sqrt(n) for n in x]

plt.plot(x, y, c = 'red')
plt.title('$y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(0, 20)
plt.ylim(0, 5);
```

```python
x = list(range(-5, 6))
y = [n ** 3 for n in x]

plt.plot(x, y, c = 'blue')
plt.title('$y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-5, 5)
plt.ylim(-150, 150);
```

```python
x = list(range(-10, 11))
y = [2 ** n for n in x]

plt.plot(x, y, c = 'green')
plt.title('$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-10, 10)
plt.ylim(0, 1000);
```

```python
x = list(range(-10, 11))
y = [1 / (n + 1) if n != -1 else None for n in x]

plt.plot(x, y, c = 'orange')
plt.title('$y = 1/x+1$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-10, 10)
plt.ylim(-1, 1);
```

# 3
Combine the figures you created in the last step into one large figure with 4 subplots.

```python
row_col = 220
plt.figure(figsize=(12, 15))

# y = sqrt(x)
plt.subplot(row_col + 1)
x = list(range(0, 21))
y = [math.sqrt(n) for n in x]

plt.plot(x, y, c = 'red')
plt.title('$y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(0, 20)
plt.ylim(0, 5)

# y = x^3
plt.subplot(row_col + 2)
x = list(range(-5, 6))
y = [n ** 3 for n in x]

plt.plot(x, y, c = 'blue')
plt.title('$y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-5, 5)
plt.ylim(-150, 150)

# y = 2^x
plt.subplot(row_col + 3)
x = list(range(-10, 11))
y = [2 ** n for n in x]

plt.plot(x, y, c = 'green')
plt.title('$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-10, 10)
plt.ylim(0, 1000)

# y = 1 / x+1
plt.subplot(row_col + 4)
x = list(range(-10, 11))
y = [1 / (n + 1) if n != -1 else None for n in x]

plt.plot(x, y, c = 'orange')
plt.title('$y = 1/x+1$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-10, 10)
plt.ylim(-1, 1);
```

# 4
Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x_min, x_max = -10, 10
y_min, y_max = -10, 10
number_of_points = 201

x = linspace(x_min, x_max, number_of_points)

equations = [
    {
        'y' : [math.sqrt(n) if n >= 0 else None for n in x],
        'c' : 'red',
        'label' : '$y = \sqrt{x}$'
    },
    {
        'y' : [n ** 3 for n in x],
        'c' : 'blue',
        'label' : '$y = x^3$'
    },
    {
        'y' : [2 ** n for n in x],
        'c' : 'green',
        'label' : '$y = 2^x$'
    },
    {
        'y' : [1 / (n + 1) if n + 1 != 0 else None for n in x],
        'c' : 'orange',
        'label' : '$y = 1/x+1$'
    }
]

for y in equations:
    plt.plot(x, y['y'], c = y['c'], label = y['label'])

plt.title('Combined Plots')
plt.xlabel('$x$')
plt.ylabel('$y$', rotation = 0)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.legend(loc = 'upper left');
```
